```python
# convert the dataframe to a list
arr = df['PC ID'].tolist()

# declare the three list to store the IDs
P_117 = [0] * 500
P_289 = [0] * 500
P_714 = [0] * 500


# split the list to two sections [department number, PC number]
character = 'P'
sliced_list = [string.split(character) for string in arr]

# loop to add the PC nmber in one of the lists (P_117, P_289, P_714)
counter = 0
for x in sliced_list:
 temp = sliced_list[counter]
 sec1 = int(temp[0])
 sec2 = int(temp[1])

 match sec1:
  case 117:
   P_117[(sec2-1)] = sec2
  case 289:
   P_289[(sec2-1)] = sec2
  case 714:
   P_714[(sec2-1)] = sec2

 counter = counter + 1
```

```python
# print the available IDs in the 117 department
print("The available IDs in 117:")
counter = 0
for x in range(max(P_117)):
  if P_117[counter] == 0:
    print('117P{}'.format(counter+1))
  counter += 1
```

```python
# print the available IDs in the 289 department
print("The available IDs in 289:")
counter = 0
for x in range(max(P_289)):
  if P_289[counter] == 0:
    print('289P{}'.format(counter+1))
  counter += 1
```

```python
# print the available IDs in the 714 department
print("The available IDs in 714:")
counter = 0
for x in range(max(P_714)):
  if P_714[counter] == 0:
    print('714P{}'.format(counter+1))
  counter += 1
```
